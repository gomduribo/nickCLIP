```python
# í•œ ê³„ì •ì˜ íŒ”ë¡œì›Œ ëª…ë‹¨ì„ ì½ì–´ì„œ ë¦¬ìŠ¤íŠ¸ë¡œ ë§Œë“¤ì–´ì¤Œ, ì›ì†ŒëŠ” dictí˜•íƒœ
f = open("./nasa_follower.txt", 'r')
data = f.read()
follower_dic_list_without=data.split("User\"}")
user_dic_list=[]
for i in range(len(follower_dic_list_without)):
    try:
        follower_dic=follower_dic_list_without[i]+"User\"}"
        # print(follower_dic)
        follower_dic=json.loads(follower_dic)
        # print(follower_dic)
        user_dic_list.append(follower_dic)
    except:
        pass
    
print(user_dic_list)
```

```python
def download(url, file_name):
    with open(file_name, "wb") as file:   # open in binary mode
        response = get(url)               # get request
        file.write(response.content)      # write to file
```

```python
# í”„ë¡œí•„ ë°°ë„ˆ ì‚¬ì§„, description, username ë°ì´í„° ì €ìž¥
count=0
for i in range(len(user_dic_list)):
    if(user_dic_list[i]['rawDescription']!=None and user_dic_list[i]['rawDescription']!="" and user_dic_list[i]['profileBannerUrl']!=None):
        print(f"nickname:{user_dic_list[i]['username']}\nthumbnail_url:{user_dic_list[i]['profileImageUrl']}\nbanner_url:{user_dic_list[i]['profileBannerUrl']}\nDescription:{user_dic_list[i]['rawDescription']}\n---------------")
        # ì´ë¯¸ì§€ ë‹¤ìš´ë¡œë“œ
        print(user_dic_list[i]['rawDescription']==None)
        download(user_dic_list[i]['profileBannerUrl'],f"/home/host_data/ThumbCrawlImage/image/{count+1}.jpg")
        
        # nickname(label) txtë¡œ ì €ìž¥
        label_txt = open(f"/home/host_data/ThumbCrawlImage/label/label_{count+1}.txt", 'w')
        label_txt.write(user_dic_list[i]['username'])

        # Description txtë¡œ ì €ìž¥
        des_txt = open(f"/home/host_data/ThumbCrawlImage/description/des_{count+1}.txt", 'w')
        des_txt.write(user_dic_list[i]['rawDescription'])
        
        count=count+1
    else: pass
    # print(f"nickname:{user_dic_list[i]['username']}\nthumbnail_url:{user_dic_list[i]['profileImageUrl']}\nbanner_url:{user_dic_list[i]['profileBannerUrl']}\nDescription:{user_dic_list[i]['rawDescription']}\n---------------")
```

```python
#ì „ì²´
crawl_users=['NASA','trvisXX','elonmusk','POTUS','neymarjr','MrBeast','ChelseaFC','PSG_inside','ChampionsLeague','TheNotoriousMMA','Stanford','MIT','Samsung','Apple','Tesla','TomHolland1996','Avengers','Hyundai_Global','McDonalds']
for i in range(len(crawl_users)):
    #í•œê³„ì •ì˜ idí™•ì¸
    os.system(f"twscrape user_by_login {crawl_users[i]} > ./crawl_list/{crawl_users[i]}.txt")
    f = open(f"./crawl_list/{crawl_users[i]}.txt", 'r')
    user_data=f.read()
    user_json=json.loads(user_data)

    #ìœ„ì—ì„œ ì–»ì€ idë¡œ ê·¸ ê³„ì •ì˜ íŒ”ë¡œì›Œ ì •ë³´ ì¶”ì¶œ
    os.system(f"twscrape followers --limit=100 {user_json['id']} > ./crawl_list/{crawl_users[i]}_follower.txt")

    # í•œ ê³„ì •ì˜ íŒ”ë¡œì›Œ ëª…ë‹¨ì„ ì½ì–´ì„œ ë¦¬ìŠ¤íŠ¸ë¡œ ë§Œë“¤ì–´ì¤Œ, ì›ì†ŒëŠ” dictí˜•íƒœ
    f2 = open(f"./crawl_list/{crawl_users[i]}_follower.txt", 'r')
    data = f2.read()
    follower_dic_list_without=data.split("User\"}")
    user_dic_list=[]
    for i in range(len(follower_dic_list_without)):
        try:
            follower_dic=follower_dic_list_without[i]+"User\"}"
            # print(follower_dic)
            follower_dic=json.loads(follower_dic)
            # print(follower_dic)
            user_dic_list.append(follower_dic)
        except:
            pass

    # í”„ë¡œí•„ ë°°ë„ˆ ì‚¬ì§„, description, username ë°ì´í„° ì €ìž¥
    count=0
    for i in range(len(user_dic_list)):
        if(user_dic_list[i]['rawDescription']!=None and user_dic_list[i]['rawDescription']!="" and user_dic_list[i]['profileBannerUrl']!=None):
            print(f"nickname:{user_dic_list[i]['username']}\nthumbnail_url:{user_dic_list[i]['profileImageUrl']}\nbanner_url:{user_dic_list[i]['profileBannerUrl']}\nDescription:{user_dic_list[i]['rawDescription']}\n---------------")
            # ì´ë¯¸ì§€ ë‹¤ìš´ë¡œë“œ
            print(user_dic_list[i]['rawDescription']==None)
            download(user_dic_list[i]['profileBannerUrl'],f"/home/host_data/ThumbCrawlImage/image/{crawl_users[i]}_{count+1}.jpg")
            
            # nickname(label) txtë¡œ ì €ìž¥
            label_txt = open(f"/home/host_data/ThumbCrawlImage/label/{crawl_users[i]}_label_{count+1}.txt", 'w')
            label_txt.write(user_dic_list[i]['username'])
    
            # Description txtë¡œ ì €ìž¥
            des_txt = open(f"/home/host_data/ThumbCrawlImage/description/{crawl_users[i]}_des_{count+1}.txt", 'w')
            des_txt.write(user_dic_list[i]['rawDescription'])
            
            count=count+1
        else: pass
        # print(f"nickname:{user_dic_list[i]['username']}\nthumbnail_url:{user_dic_list[i]['profileImageUrl']}\nbanner_url:{user_dic_list[i]['profileBannerUrl']}\nDescription:{user_dic_list[i]['rawDescription']}\n---------------")
```
