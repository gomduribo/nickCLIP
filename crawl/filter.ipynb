```python
import os
```

```python
crawl_folder = 'ThumbCrawlImage'
description_folder = 'description'
image_folder = 'image'
label_folder = 'label'

description_path = os.path.join(crawl_folder, description_folder)
image_path = os.path.join(crawl_folder, image_folder)
label_path = os.path.join(crawl_folder, label_folder)

description_files = os.listdir(description_path)
image_files = os.listdir(image_path)
label_files = os.listdir(label_path)
```

```python
filtered_description_folder = 'filtered/description/'
filtered_image_folder = 'filtered/image/'
filtered_label_folder = 'filtered/label/'
```

```python
word = 'A'
ascii = ord(word)
print(ascii)
word = 'z'
print(ord(word))
    
```

```python
import shutil

description_files = os.listdir(description_path)
# description 파일들을 읽어서 영문자만 남기고 새로운 파일로 저장

for file in description_files:
    # 새로 작성할 line 저장
    new_line = ''
    lines = open(os.path.join(description_path, file), 'r').readlines()
    for line in lines:
        for word in line:
            ascii = ord(word)
            # 아스코드값으로 영문자인지 확인 32번은 공백
            if ascii >= 65 and ascii <= 122 or ascii == 32:
                new_line += word
    filtered_description_file = open(os.path.join(filtered_description_folder, file), 'w')
    filtered_description_file.write(new_line)
    filtered_description_file.close()


```

```python
#filter에 이미지, 라벨 복사
for file in image_files:
    shutil.copy(os.path.join(image_path, file), os.path.join(filtered_image_folder, file))
for file in label_files:
    shutil.copy(os.path.join(label_path, file), os.path.join(filtered_label_folder, file))
```

```python

#  filtered description 파일이 10글자보다 적으면 삭제

filtered_description_files = os.listdir(filtered_description_folder)
for file in filtered_description_files:
    lines = open(os.path.join(filtered_description_folder, file), 'r').readlines()
    description_file = file
    label_file = file
    image_file = file.replace('.txt', '.jpg')
    print(file)
    if lines == None:
        os.remove(os.path.join(filtered_description_folder, description_file))
        os.remove(os.path.join(filtered_image_folder, image_file))
        os.remove(os.path.join(filtered_label_folder, label_file)) 
    else:
        line_length = 0
        for line in lines:
            line_length += len(line)
        if line_length < 10:
            os.remove(os.path.join(filtered_description_folder, description_file))
            os.remove(os.path.join(filtered_image_folder, image_file))
            os.remove(os.path.join(filtered_label_folder, label_file))
```

```python
# 닉네임이 영어가 아니면 삭제
filtered_label_files = os.listdir(filtered_label_folder)
for file in filtered_label_files:
    lines = open(os.path.join(filtered_label_folder, file), 'r').readlines()
    description_file = file
    label_file = file
    image_file = file.replace('.txt', '.jpg')
    for line in lines:
        for word in line:
            ascii = ord(word)
            if ascii < 65 or ascii > 122 :
                os.remove(os.path.join(filtered_description_folder, file))
                os.remove(os.path.join(filtered_image_folder, image_file))
                os.remove(os.path.join(filtered_label_folder, label_file))
                break
```

```python
# 사진을 다운받고나서 사진이 없는 경우가 있음. 그런 경우 삭제
filtered_image_files = os.listdir(filtered_image_folder)
for file in filtered_image_files:
    
    description_file = file.replace('.jpg', '.txt')
    label_file = file.replace('.jpg', '.txt')
    image_file = file
    if os.path.getsize(os.path.join(filtered_image_folder, image_file)) <= 1:
        os.remove(os.path.join(filtered_description_folder, description_file))
        os.remove(os.path.join(filtered_image_folder, image_file))
        os.remove(os.path.join(filtered_label_folder, label_file))
            
```
