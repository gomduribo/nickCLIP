```python
import os
from requests import get
import json
```

```python
def download(url, file_name):
    with open(file_name, "wb") as file:   # open in binary mode
        response = get(url)               # get request
        file.write(response.content)      # write to file
```

```python
#전체
# crawl_users=['NASA','trvisXX','elonmusk','POTUS','neymarjr','MrBeast','ChelseaFC','PSG_inside','ChampionsLeague','TheNotoriousMMA','Stanford','MIT','Samsung','Apple','Tesla','TomHolland1996','Avengers','Hyundai_Global','McDonalds']
crawl_users=['KMbappe','Cristiano','TheSunFootball','BurgerKing','NFL','NewJeans_ADOR','jimmykimmel','USArmy','premierleague','OpenAI','BrunoMars','nypost','PostMalone','Yankees','Rangers']
for i in range(len(crawl_users)):
    #한계정의 id확인
    os.system(f"twscrape user_by_login {crawl_users[i]} > ./crawl_list/{crawl_users[i]}.txt")
    f = open(f"./crawl_list/{crawl_users[i]}.txt", 'r')
    user_data=f.read()
    user_json=json.loads(user_data)

    #위에서 얻은 id로 그 계정의 팔로워 정보 추출
    os.system(f"twscrape followers --limit=100 {user_json['id']} > ./crawl_list/{crawl_users[i]}_follower.txt")

    # 한 계정의 팔로워 명단을 읽어서 리스트로 만들어줌, 원소는 dict형태
    f2 = open(f"./crawl_list/{crawl_users[i]}_follower.txt", 'r')
    data = f2.read()
    follower_dic_list_without=data.split("User\"}")
    user_dic_list=[]
    for k in range(len(follower_dic_list_without)):
        try:
            follower_dic=follower_dic_list_without[k]+"User\"}"
            # print(follower_dic)
            follower_dic=json.loads(follower_dic)
            # print(follower_dic)
            user_dic_list.append(follower_dic)
        except:
            pass

    # 프로필 배너 사진, description, username 데이터 저장
    count=0
    for j in range(len(user_dic_list)):
        if(user_dic_list[j]['rawDescription']!=None and user_dic_list[j]['rawDescription']!="" and user_dic_list[j]['profileBannerUrl']!=None):
            print(f"nickname:{user_dic_list[j]['username']}\nthumbnail_url:{user_dic_list[j]['profileImageUrl']}\nbanner_url:{user_dic_list[j]['profileBannerUrl']}\nDescription:{user_dic_list[j]['rawDescription']}\n---------------")
            # 이미지 다운로드
            print(user_dic_list[j]['rawDescription']==None)
            download(user_dic_list[j]['profileBannerUrl'],f"/home/host_data/ThumbCrawlImage/image/{crawl_users[i]}_{count+1}.jpg")
            
            # nickname(label) txt로 저장
            label_txt = open(f"/home/host_data/ThumbCrawlImage/label/{crawl_users[i]}_label_{count+1}.txt", 'w')
            label_txt.write(user_dic_list[j]['username'])
    
            # Description txt로 저장
            des_txt = open(f"/home/host_data/ThumbCrawlImage/description/{crawl_users[i]}_des_{count+1}.txt", 'w')
            des_txt.write(user_dic_list[j]['rawDescription'])
            
            count=count+1
        else: pass
        # print(f"nickname:{user_dic_list[i]['username']}\nthumbnail_url:{user_dic_list[i]['profileImageUrl']}\nbanner_url:{user_dic_list[i]['profileBannerUrl']}\nDescription:{user_dic_list[i]['rawDescription']}\n---------------")
```
