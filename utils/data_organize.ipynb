# 모은 데이터를 train/val/test로 분할

```python
import os
import shutil
```

```python
img_path="/home/host_data/ThumbCrawlImage/image/"
label_path="/home/host_data/ThumbCrawlImage/label/"
des_path="/home/host_data/ThumbCrawlImage/description/"
image_name_list=os.listdir(img_path)
label_name_list=os.listdir(label_path)
des_name_list=os.listdir(des_path)
```

```python

train_num=100
val_num=20
test_num=10

dst_img_train_path='/home/host_data/nickData/image/train/'
dst_label_train_path='/home/host_data/nickData/label/train/'
dst_des_train_path='/home/host_data/nickData/description/train/'

dst_img_val_path='/home/host_data/nickData/image/val/'
dst_label_val_path='/home/host_data/nickData/label/val/'
dst_des_val_path='/home/host_data/nickData/description/val/'

dst_img_test_path='/home/host_data/nickData/image/test/'
dst_label_test_path='/home/host_data/nickData/label/test/'
dst_des_test_path='/home/host_data/nickData/description/test/'

#train
for i in range(0,train_num):
    # 이미지 이동
    shutil.copy(img_path+image_name_list[i], dst_img_train_path+image_name_list[i])

    # 라벨 이동
    label_name=image_name_list[i].split(".jpg")[0]+"_label.txt"
    shutil.copy(label_path+label_name, dst_label_train_path+label_name)
    
    # des 이동
    des_name=image_name_list[i].split(".jpg")[0]+"_des.txt"
    shutil.copy(des_path+des_name, dst_des_train_path+des_name)
    print(f"{image_name_list[i]} {label_name} {des_name}\n\n")

#val
for j in range(train_num, train_num+val_num):
    # 이미지 이동
    shutil.copy(img_path+image_name_list[j], dst_img_val_path+image_name_list[j])

    # 라벨 이동
    label_name=image_name_list[j].split(".jpg")[0]+"_label.txt"
    shutil.copy(label_path+label_name, dst_label_val_path+label_name)
    
    # des 이동
    des_name=image_name_list[j].split(".jpg")[0]+"_des.txt"
    shutil.copy(des_path+des_name, dst_des_val_path+des_name)
    print(f"{image_name_list[j]} {label_name} {des_name}\n\n")

#test
for k in range(train_num+val_num, train_num+val_num+test_num):
    # 이미지 이동
    shutil.copy(img_path+image_name_list[k], dst_img_test_path+image_name_list[k])

    # 라벨 이동
    label_name=image_name_list[k].split(".jpg")[0]+"_label.txt"
    shutil.copy(label_path+label_name, dst_label_test_path+label_name)
    
    # des 이동
    des_name=image_name_list[k].split(".jpg")[0]+"_des.txt"
    shutil.copy(des_path+des_name, dst_des_test_path+des_name)
    print(f"{image_name_list[k]} {label_name} {des_name}\n\n")
    
```
