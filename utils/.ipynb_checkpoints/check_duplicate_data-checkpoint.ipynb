```python
import os 
```

```python
train_img_list=os.listdir('/workspace/team2/data/filter_50000/train/image/')
valid_img_list=os.listdir('/workspace/team2/data/filter_50000/valid/image/')
```

```python
count=0
dup_list=[]
for train_file in train_img_list:
    if(train_file in valid_img_list):
        count=count+1
        dup_list.append(train_file)
        print(f"{count}-{train_file}")
```
